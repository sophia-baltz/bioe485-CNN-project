```python
from PIL import Image
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(".\BOSQUE_files\BOSQUE_metadata.csv")
df.head()
```

```python
total_imgs = len(df)

labels = np.empty(total_imgs)
lesions = np.empty((total_imgs,224,224,3))
skin_type = np.empty(total_imgs)



for row in df.itertuples():
    i = row.Index
    #labeling 0 for benign, 1 for pre-cancerous or cancerous
    les_type = row.lesion_nature
    if les_type == "malignant":
        labels[i] = 1
    else:
        labels[i] = 0

    
    file = row.file_name
    path = ".\BOSQUE_files\BOSQUE_test_set\\" + file + ".jpg"
    img = Image.open(path)
    img = img.resize((224,224))

    #TODO: potential pre-processing here. randomly rotate img & mutate colors
    np_img = np.asarray(img)
    #divide by 225 so all values are between 0 and 1
    lesions[i] = np_img / 225

    skin = row.skin_phototype
    if skin == "II": 
        skin_type[i] = 2
    elif skin == "III": 
        skin_type[i] = 3
    elif skin == "IV": 
        skin_type[i] = 4
    elif skin == "V": 
        skin_type[i] = 5   
    elif skin == "VI": 
        skin_type[i] = 6 
```

```python
data = np.empty(3, dtype=object)
data[0] = lesions
data[1] = labels
data[2] = skin_type
np.save('BOSQUE_data',data)
```
