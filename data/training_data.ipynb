```python
import kagglehub
from kagglehub import KaggleDatasetAdapter
from PIL import Image
import numpy as np
```

```python
# Set the path to the file you'd like to load
file_path = "HAM10000_metadata.csv"

# Load the latest version

metadata = kagglehub.dataset_load(
  KaggleDatasetAdapter.PANDAS,
  "kmader/skin-cancer-mnist-ham10000",
  file_path,
  # Provide any additional arguments like 
  # sql_query or pandas_kwargs. See the 
  # documenation for more information:
  # https://github.com/Kaggle/kagglehub/blob/main/README.md#kaggledatasetadapterpandas
)

print("First 5 records:", metadata.head())
total_imgs = metadata.shape[0]
```

```python
labels = np.empty(total_imgs)
lesions = np.empty((total_imgs,224,224,3))
#benign: bkl,df,nv,vasc
#malignant: akiec,bcc,mel
for i in range(total_imgs):
    #labeling 0 for benign, 1 for pre-cancerous or cancerous
    les_type = metadata.at[i,"dx"]
    if les_type == "akiec" or les_type == "bcc" or les_type == "mel":
        labels[i] = 1
    else:
        labels[i] = 0

    
    file = metadata.at[i,"image_id"]
    #CHANGE HERE: add path to file images in local computer 
    path = "C:\\Users\\sylvi\\OneDrive - University of Illinois - Urbana\\Desktop\\Classes\\BIOE485-mathofmachinelearning\\HAM10000_images\\" + file + ".jpg"
    img = Image.open(path)
    img = img.resize((224,224))
    #TODO: potential pre-processing here. randomly rotate img & mutate colors
    np_img = np.asarray(img)
    #divide by 225 so all values are between 0 and 1
    lesions[i] = np_img / 225    
```

```python
#save files to use in model
np.save('HAM_labels', labels)
np.save('HAM_lesions',lesions)
```
